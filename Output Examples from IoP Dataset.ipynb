```python
import nibabel as nib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy.misc
import scipy.ndimage
from scipy.stats import norm

import os

from skimage.measure import compare_ssim as cssim
from skimage.measure import compare_psnr as cpsnr
```

```python
f, axarr = plt.subplots(1,3, figsize=(20,20))

img = nib.load('D:\IXI-T1-IOP\IXI306-IOP-0867-T1.nii.gz')
data1 = np.rot90(img.get_data())
axarr[0].imshow(data1[:,:,72], cmap = 'gray', vmin=np.amin(data1[:,:,72]), vmax=np.amax(data1[:,:,72]))

img = nib.load('D:\IXI-T1-IOP-DEGRADED\IXI306-IOP-0867-T1.nii.gz')
data2 = np.rot90(img.get_data())
axarr[1].imshow(data2[:,:,72], cmap = 'gray', vmin=np.amin(data2[:,:,72]), vmax=np.amax(data2[:,:,72]))

img = nib.load('D:\Image1\IXI306IOP0867T1_niftynet_out.nii.gz')
data3 = np.rot90(img.get_data())
plt.imshow(data3[:,:,72,0,0], cmap = 'gray', vmin = np.amin(data3[:,:,72,0,0]), vmax = np.amax(data3[:,:,72,0,0]))

print(cpsnr(data1.astype(np.float32),data3[:,:,:,0,0],data_range=np.amax(data1)-np.amin(data1)))
print(cssim(data1.astype(np.float32),data3[:,:,:,0,0],data_range=np.amax(data1)-np.amin(data1)))
```
