```python
f, axarr = plt.subplots(2,2, figsize=(10,10))

img = nib.load('D:/IXI-T1-IOP/IXI306-IOP-0867-T1.nii.gz')
data1 = np.rot90(img.get_data()).astype(np.float64)
axarr[0,0].imshow(data1[:,:,72], cmap = 'gray', vmin=np.amin(data1[:,:,72]), vmax=np.amax(data1[:,:,72]))

img = nib.load('D:/IXI-T1-IOP-DEGRADED/IXI306-IOP-0867-T1.nii.gz')
data2 = np.rot90(img.get_data()).astype(np.float64)
axarr[0,1].imshow(data2[:,:,36], cmap = 'gray', vmin=np.amin(data2[:,:,36]), vmax=np.amax(data2[:,:,36]))

img = nib.load('D:/IXI-T1-IOP-INTERP/IXI306-IOP-0867-T1.nii.gz')
data3 = np.rot90(img.get_data()).astype(np.float64)
axarr[1,0].imshow(data3[:,:,72], cmap = 'gray', vmin=np.amin(data3[:,:,72]), vmax=np.amax(data3[:,:,72]))

# case of overfitting
img = nib.load('D:/IXI-T1-SR-1/IXI306IOP0867T1_niftynet_out.nii.gz')
data4 = np.rot90(img.get_data()).astype(np.float64)
axarr[1,1].imshow(data4[:,:,72,0,0], cmap = 'gray', vmin = np.amin(data4[:,:,72,0,0]), vmax = np.amax(data4[:,:,72,0,0]))
```
