```python
# Reg_aladin tool (reg_aladin.exe) [1] requires building NiftyReg [2] with cmake in the case of Windows.

# [1] http://cmictig.cs.ucl.ac.uk/wiki/index.php/Reg_aladin
# [2] http://cmictig.cs.ucl.ac.uk/wiki/index.php/NiftyReg
```

```python
import os
import subprocess
```

```python
input_path = 'D:/IXI-T1/'
f_T1 = []
for root, dirs, filenames in os.walk(input_path):
    for f in filenames:
        fullpath = os.path.join(input_path, f)
        f_T1.append(fullpath)
```

```python
input_path = 'D:/IXI-T2/'
f_T2 = []
for root, dirs, filenames in os.walk(input_path):
    for f in filenames:
        fullpath = os.path.join(input_path, f)
        f_T2.append(fullpath)
```

```python
f_RES = []
for k in range(0,577):
    f_RES.append('D:/DataReg/IXI-T2-RES/'+f_T2[k][10:])
```

```python
for k in range(0,2):
    print(k)
    subprocess.call(['D:/myNiftyReg/bin/reg_aladin.exe', '-ref', f_T1[k], '-flo', f_T2[k], '-rigOnly', '-res', f_RES[k]])
```

```python
input_path = 'D:/IXI-PD/'
f_PD = []
for root, dirs, filenames in os.walk(input_path):
    for f in filenames:
        fullpath = os.path.join(input_path, f)
        f_PD.append(fullpath)
```

```python
f_RES = []
for k in range(0,577):
    f_RES.append('D:/DataReg/IXI-PD-RES/'+f_PD[k][10:])
```

```python
for k in range(0,2):
    print(k)
    subprocess.call(['D:/myNiftyReg/bin/reg_aladin.exe', '-ref', f_T1[k], '-flo', f_PD[k], '-rigOnly', '-res', f_RES[k]])
```
