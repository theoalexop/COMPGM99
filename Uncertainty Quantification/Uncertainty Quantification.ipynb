```python
import nibabel as nib
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
input_path_dropout_samples = 'D:/IXI-T1-DROPOUT-SAMPLES/'
data = []

for root, dirs, filenames in os.walk(input_path_dropout_samples):
    for f in filenames:
        print(f)
        fullpath = os.path.join(input_path_dropout_samples, f)
        img = nib.load(fullpath)
        data.append(img.get_data().astype(np.float64))

data2 = np.asarray(data)        

# Computation of per voxel median.
median = np.median(data2, axis = 0)

f, axarr = plt.subplots(1,2, figsize=(10,10))
axarr[0].imshow(np.rot90(median[:,:,72,0,0]), cmap = 'gray', vmin = np.amin(median[:,:,72,0,0]), vmax = np.amax(median[:,:,72,0,0]))

# Computation of per voxel standard deviation.
stdev = np.std(data2, axis = 0)

axarr[1].imshow(np.rot90(stdev[:,:,72,0,0]), cmap = 'gray', vmin = np.amin(stdev[:,:,72,0,0]), vmax = 0.4*np.amax(stdev[:,:,72,0,0]))
plt.savefig('D:/T1-IOP-median-std.png',bbox_inches='tight')
```
